```python
import requests, bs4, re
import pandas as pd
```

```python
query = 'for loop java'
url = 'https://stackoverflow.com/search?q='+query
```

```python
def vote_ans(child1):
    nums = re.findall(r'\d+', child1)
    if (len(nums)) == 2:
        return nums[0], nums[1]
    elif (len(nums)) == 1:
        return nums[0], 0
    else:
        return 0, 0
```

```python
def link(child3):
    return 'https://stackoverflow.com'+re.findall(r'href="(.*?)"', child3)[0]
def title(child3):
    return re.findall(r'<a .*? title="(.*?)">', child3)[0]
def summary(child3):
    raw = re.findall(r'<div class="excerpt">((.|\n)*?)</div>', child3)[0][0]
    raw = raw.strip()
    remove_tags_re = re.compile(r"<.*?>")
    raw = remove_tags_re.sub("", raw)
    raw = raw.replace("&lt;", "<")
    raw = raw.replace("&gt;", ">")
    raw = raw.replace("&amp;", "&")
#     raw = raw.replace("<span (.*)?>", ' ')
    return raw
```

```python
child3 = """<div class="summary">
<div class="result-link">
<h3>
<a class="question-hyperlink" data-searchsession="/questions/85190/how-does-the-java-for-each-loop-work?r=SearchResults&amp;s=2|286.2318" href="/questions/85190/how-does-the-java-for-each-loop-work?r=SearchResults" title="How does the Java 'for each' loop work?">
Q: How does the Java 'for each' loop work?        </a>
</h3>
</div>
<div class="excerpt">
                Consider:

List&lt;String&gt; someList = new ArrayList&lt;String&gt;();
// add "monkey", "donkey", "skeleton key" to someList




<span class="result-highlight">for</span> (String item : someList) {
    System.out.println(item);
}


What would the equivalent <span class="result-highlight">for</span> <span class="result-highlight">loop</span> look like without using the <span class="result-highlight">for</span> each syntax?
 … 
            </div>
<div class="tags user-tags t-java t-foreach t-syntactic-sugar">
<a class="post-tag" href="/questions/tagged/java" rel="tag" title="show questions tagged 'java'">java</a> <a class="post-tag" href="/questions/tagged/foreach" rel="tag" title="show questions tagged 'foreach'">foreach</a> <a class="post-tag" href="/questions/tagged/syntactic-sugar" rel="tag" title="show questions tagged 'syntactic-sugar'">syntactic-sugar</a> </div>
<div class="started fr">
asked <span class="relativetime" title="2008-09-17 16:44:32Z">Sep 17 '08</span> by <a href="/users/5074/jay-r">Jay R.</a> </div>
</div>"""

raw = re.findall(r'<div class="excerpt">((.|\n)*?)</div>', child3)[0][0]
raw = raw.strip()
raw = raw.replace("&lt;", "<")
raw = raw.replace("&gt;", ">")
raw = raw.replace("&amp;", "&")
remove_tags_re = re.compile(r"<.*?>")
raw = remove_tags_re.sub("", raw)
print(raw)
```

```python
results = pd.DataFrame(columns=['Title', 'Link', 'Summary', 'Vote', 'Answers'])
res = requests.get(url)
if res.status_code == requests.codes.ok:
    ressoup = bs4.BeautifulSoup(res.text, 'lxml')
    elems = ressoup.select('.question-summary')
    for i,e in enumerate(elems):
        print(i)
        for j, c in enumerate(e.children):
            if j == 1:
                vote, ans = vote_ans(str(c))
                print(vote, ans)
            if j == 3:
                t = title(str(c))
                print(t)
                l = link(str(c))
                print(l)
                s = summary(str(c))
                print(s)
        results = pd.concat([results, pd.DataFrame({
            'Title': [t], 'Link': [l], 'Summary': [s], 'Vote': [vote], 'Answers': [ans]
        })], axis=0)
        print('-------')
else:
    print('Something went wrong')
```

```python
results= results.reset_index(drop=True)
results
```

```python
results['Excerpt'] = results['Title'] +'\n\n' + results['Summary']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(results['Summary'])
vect = vectorizer.transform(['for loop java'])
```

```python
results['Similarity'] = results['Summary'].apply(lambda x : cosine_similarity(vectorizer.transform([x]),vect)[0][0])
```
