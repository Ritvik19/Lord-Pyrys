# Libraries

```python
import requests, bs4
import numpy as np
import pandas as pd
from IPython.display import clear_output

pd.options.display.max_columns = 5
pd.options.display.max_colwidth = -1
```

# Scraping

```python
URL = 'https://jaxenter.com/15-useful-code-snippets-java-developers-131796.html'
res = requests.get(URL)
if res.status_code == requests.codes.ok:
#     print(res.text)
    ressoup = bs4.BeautifulSoup(res.text, 'lxml')
    elems_h2 = ressoup.select('h2')
    elems_code = ressoup.select('pre')
    h2 = []
    pre = []
    for e1 in elems_h2:
        h2.append(e1.getText())
    for e2 in elems_code:
        pre.append(e2.getText())
    pre[1] = pre[1] + '\n\n' + pre[2]
    pre.pop(2)
else:
    print('Something went wrong')
```

```python
Java = pd.DataFrame({
    'Title': h2,
    'Code': pre
})
Java
```
