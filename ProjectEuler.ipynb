# Libraries

```python
import pandas as pd
import requests, bs4
from IPython.display import clear_output
from time import sleep

pd.options.display.max_columns = 5
pd.options.display.max_colwidth = -1
```

# Scraping

```python
titles = []
problems = []
```

```python
for i in range(1,663):
    print(i, '/662')
    url = 'https://projecteuler.net/problem='+str(i)
    res = requests.get(url)
    if res.status_code == requests.codes.ok:
        ressoup = bs4.BeautifulSoup(res.text, 'lxml')
        title = ressoup.select('h2')[0].getText()
        titles.append(title)
        print(title)
        problem = ressoup.select('.problem_content')[0].getText()
        problems.append(problem)
        print(problem)
        sleep(1)
        clear_output(wait=True)
    else:
        print('Something went wrong')
```

```python
ProjectEuler = pd.DataFrame()
ProjectEuler['Title'] = titles
ProjectEuler['Problem'] = problems
```
